```python
import pandas as pd
import sys

accident = pd.read_csv('dftRoadSafety_Accidents_2016.csv')
casualty = pd.read_csv('Cas.csv')
vehicle = pd.read_csv('MakeModel2016.csv')

pd.set_option('display.max_columns', 70)
```

```python
main = pd.merge(accident, casualty, on=['Accident_Index'])
main.head()
```

```python
major = pd.merge(main, vehicle, on=['Accident_Index'])
major.head()
```

```python
def quality_report(df):
    """
    Description: Displays quality of data in terms of missing values, unique numbers, datatypes etc.
    Arguments: Dataframe
    """
    nuniq = df.T.apply(lambda x: x.nunique(), axis=1)
    total = df.isnull().sum().sort_values(ascending = False)
    percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
    quality_df  = pd.concat([total, percent, nuniq], axis=1, keys=['Total', 'Percent','Nunique'])
    display(quality_df)
```

## Question 4

### Are certain cars more accident prone than the others

```python
car_prone = major[['Vehicle_Type', 'Age_of_Vehicle','make','model', 'Number_of_Casualties', 'Sex_of_Driver', 'Propulsion_Code']]
```

```python
car_prone.head()
```

```python
car_prone = car_prone.loc[(car_prone['Vehicle_Type'].isin([8,9])) & (car_prone['make'])]
#where 8 and 9 indicate car vehicle type
```

```python
quality_report(car_prone)
```

```python
car_prone.corr()
```

```python
### Visualizing Correlation
sns.heatmap(car_prone.corr())
```

```python
## Encoding Proportion_Code
dict_ = {1:'Petrol', 2: 'Heavy oil', 3: 'Electric', 5:'Gas', 6:'Petrol/Gas(LPG)',7: 'Gas/Bi-fuel', 8: 'Hybrid electric', 10:'New fuel technology', 12:'Electric diesel', -1:'Petrol'}
car_prone['Propulsion_Code'].replace(dict_, inplace = True)
```

```python
### Checking percentage of Car make
car_prone['make'].value_counts(normalize =True)*100
```

### Visualization of Car make versus Number of Casualty

```python
plt.figure(figsize = (20, 45))
plt.tight_layout()
plt.xlabel('Number_of_Casualties', fontsize =20)
plt.ylabel('make', fontsize =20)
car_prone.groupby('make')['Propulsion_Code'].count().sort_values(ascending=True).plot(kind='barh')
```

```python
car_prone.groupby('make')['Number_of_Casualties'].count().sort_values(ascending= False)
```

### Taking a closer look at Cars and visualizing cars with more than 1% make in the country

```python
final = car_prone.loc[(car_prone['make'].isin(['FORD','VAUXHALL','VOLKSWAGEN','PEUGEOT','BMW','TOYOTA','RENAULT', 'AUDI', 'NISSAN', 'MERCEDES', 'CITROEN', 'HONDA', 'FIAT', 'SKODA', 'SEAT', 'HYUNDAI', 'MINI', 'LAND ROVER', 'KAI', 'MAZDA', 'VOLVO', 'SUZUKI']))]
```

```python
final['make'].value_counts(normalize =True)*100
```

```python
final.groupby('make')['Number_of_Casualties'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize = (20,12))
plt.tight_layout()
plt.title('Car_Make vs Number_of_Casualties', fontsize =20)
plt.xlabel('Number_of_Casualties', fontsize =20)
plt.ylabel('make', fontsize =20)
final.groupby('make')['Number_of_Casualties'].count().sort_values(ascending=True).plot(kind='barh')
```
